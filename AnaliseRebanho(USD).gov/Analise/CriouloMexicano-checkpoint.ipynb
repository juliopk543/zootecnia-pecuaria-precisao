```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
CattleWeight = pd.read_csv(r"C:\Users\julio\Desktop\intergado\DB\.ipynb_checkpoints\CattleWeight.csv")
cattleMC = CattleWeight[CattleWeight.breed == 'MC']
```

```python
# figure
fig, axes = plt.subplots(1,2, figsize=(22, 5))

# Title
fig.suptitle('(~ Iniciando o processo de analise ~)', fontsize=22, y=1.10)

# Plot (boxplot) 0
sns.boxplot(ax=axes[0], data=cattleMC, x='month', y='weight', color='#9b59b6')
axes[0].set_title('Variao de Peso por Raa durante os meses do Ano (2015 - 2016)', fontsize=16, color='rebeccapurple')

# Plot (text) 1
axes[1].axis("off")  
axes[1].text(-0.01, 0.75,                                                               
         'A Raa Crioulo Mexicano se apresentou com \
             \n a maior volatilidade no peso',  
         fontsize=20,                                                             
         color='black',                                                         
         weight="bold");

# Quantidade de gados
qtd = len(cattleMC.cow_id.unique())
axes[1].text(0.1, 0.48, f'$\\bfSo$ $\\bf{qtd}$ $\\bfGados$',
        horizontalalignment='left',
        verticalalignment='center',
        fontsize=70, color='#400080',
       );

# Desvio padro relativo / coeficiente de variao
RSD = round((cattleMC['weight'].std() * 100 ) / cattleMC['weight'].mean(), 2)
axes[1].axis("off")  
axes[1].text(-0.01, 0.15,                                                               
         f' Com o desvio padro relativo de {RSD} %, sobre a media.',  
         fontsize=20,                                                             
         color='black',                                                         
         weight="bold");
```

```python
# Ignore Warning Setting With Copy
import warnings
from pandas.core.common import SettingWithCopyWarning
warnings.simplefilter(action='ignore', category=SettingWithCopyWarning)

# Convertendo Arroba
cattleMC['arroba'] = (cattleMC.weight / 15)
cattleMC['arroba'] = pd.cut(x=cattleMC['arroba'].astype(int), bins=[1, 11, 16, 20, 23, 26],
                    labels=['10 - 11', '12 - 15', '16 - 19',
                            '20 - 22', '23 - 25'])
```

```python
textcolor = '#2f4f4f'
palettecolor = ['#CDCDCD','#9b59b6']
# figure
fig, axes = plt.subplots(2,2, figsize=(22, 10))

# title
fig.suptitle('(~ Analise de influencia do tratamento cohort ~)', fontsize=22, y=1.02, color=textcolor)

# plot (barplot)[arroba]
sns.barplot(ax=axes[0,0], data=cattleMC, x="arroba", y="weight", hue="cohort", palette=palettecolor)
axes[0,0].set_title('Variao de peso em arroba por tipo de tratamento', fontsize=16, color=textcolor, weight="bold")

# plot (barplot)[lastWeight]
lastweight = cattleMC.groupby(['month','cow_id', 'cohort'])['weight'].last().reset_index().sort_values(by=['weight'])
sns.barplot(ax=axes[0,1], data=lastweight, x="month", y="weight", hue="cohort", palette=palettecolor)
axes[0,1].set_title('Peso de abate final por ms, por tipo de tratamento', fontsize=16, color=textcolor,weight="bold")

# Plot (text)
C2 = cattleMC[cattleMC.cohort == 'C2']['weight'].max()
C1 = cattleMC[cattleMC.cohort == 'C1']['weight'].max()

WEIGHTMAX = round((C2 * 100)/ C1, 2)

axes[1,0].text(0.45, 0.65,                                                               
         f'Podemos avaliar que o metodo cohort foi divido em C1 - C2 \
             \nonde C2 alcanou  $\\bf{WEIGHTMAX}$% do peso total de C1.',  
         fontsize=17,                                                             
         color=textcolor,                                                         
         weight="bold");

axes[1,0].axis("off")  
axes[1,0].text(0.45, 0.45,                                                               
         'O cohort pode ser interpretado como um modelo de tratamento,\
             \num plano que foi seguido, porem aqui no h informaes sobre o metodo utilizado.',  
         fontsize=17,                                                             
         color=textcolor,                                                         
         weight="bold"); 

axes[-1, -1].axis('off');
```
