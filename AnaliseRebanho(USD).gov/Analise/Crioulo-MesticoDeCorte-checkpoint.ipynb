```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
CattleWeight = pd.read_csv(r"C:\Users\julio\Desktop\intergado\DB\.ipynb_checkpoints\CattleWeight.csv")
cattleXC = CattleWeight[CattleWeight.breed == 'XC']
```

```python
# Outline identificado. 2 Gados com 0.0KG

outline = list(cattleXC[cattleXC.arroba == '0-11'].cow_id.unique())
cattleXC = cattleXC.query(f'cow_id not in {outline}')
```

```python
# figure
fig, axes = plt.subplots(1,2, figsize=(22, 5))

# Title
fig.suptitle('(~ Iniciando o processo de analise ~)', fontsize=22, y=1.10)

# Plot (boxplot) 0
sns.boxplot(ax=axes[0], data=cattleXC, x='month', y='weight', color='#3498db')
axes[0].set_title('Variao de Peso por Raa durante os meses do Ano (2015 - 2016)', fontsize=16, color='black')

# Plot (text) 1
axes[1].axis("off")  
axes[1].text(-0.01, 0.75,                                                               
         'A Raa Crioulo  Mestio de Corte se apresentou com \
             \n a maior volatilidade no peso',  
         fontsize=20,                                                             
         color='black',                                                         
         weight="bold");

# Quantidade de gados
qtd = len(cattleXC.cow_id.unique())
axes[1].text(0.1, 0.48, f'$\\bfSo$ $\\bf{qtd}$ $\\bfGados$',
        horizontalalignment='left',
        verticalalignment='center',
        fontsize=70, color='#3498db',
       );

# Desvio padro relativo / coeficiente de variao
RSD = round((cattleXC['weight'].std() * 100 ) / cattleXC['weight'].mean(), 2)
axes[1].axis("off")  
axes[1].text(-0.01, 0.15,                                                               
         f' Com o desvio padro relativo de {RSD} %, sobre a media.',  
         fontsize=20,                                                             
         color='black',                                                         
         weight="bold");
```

```python
textcolor = '#2f4f4f'
palettecolor = ['#CDCDCD','#3498db']

# figure
fig, axes = plt.subplots(2,2, figsize=(22, 10))

# title
fig.suptitle('(~ Analise de influencia do tratamento cohort ~)', fontsize=22, y=1.02, color=textcolor)

# plot (barplot)[arroba]
sns.barplot(ax=axes[0,0], data=cattleXC, x="arroba", y="weight", hue="cohort", palette=palettecolor)
axes[0,0].set_title('Variao de peso em arroba por tipo de tratamento', fontsize=16, color=textcolor, weight="bold")

# plot (barplot)[lastWeight]
lastweight = cattleXC.groupby(['month','cow_id', 'cohort'])['weight'].last().reset_index().sort_values(by=['weight'])
sns.barplot(ax=axes[0,1], data=lastweight, x="month", y="weight", hue="cohort", palette=palettecolor)
axes[0,1].set_title('Peso de abate final por ms, por tipo de tratamento', fontsize=16, color=textcolor,weight="bold")

# Plot (text)
C2 = cattleXC[cattleXC.cohort == 'C2']['weight'].max()
C1 = cattleXC[cattleXC.cohort == 'C1']['weight'].max()

WEIGHTMAX = round((C2 * 100)/ C1, 2)

axes[1,0].text(0.45, 0.65,                                                               
         f'Podemos avaliar que o metodo cohort foi divido em C1 - C2 \
             \nonde C2 alcanou  $\\bf{WEIGHTMAX}$% do peso total de C1.',  
         fontsize=17,                                                             
         color=textcolor,                                                         
         weight="bold");

axes[1,0].axis("off")  
axes[1,0].text(0.45, 0.45,                                                               
         'O cohort pode ser interpretado como um modelo de tratamento,\
             \num plano que foi seguido, porem aqui no h informaes sobre o metodo utilizado.',  
         fontsize=17,                                                             
         color=textcolor,                                                         
         weight="bold"); 

axes[-1, -1].axis('off');
```

```python
cattleXC_pesoMensal = cattleXC.groupby(by=['month', 'year', 'cow_id', 'cohort'])['weight'].max().reset_index().sort_values(by='cow_id').rename(columns={'weight': 'weightMAX'})
```

```python
PesoRebanho_janeiro = cattleXC_pesoMensal[cattleXC_pesoMensal.month == 'January']['weightMAX'].mean()

# Peso Final Esperado
PesoFinalEsperado = (PesoRebanho_janeiro + 90) # https://dicas.boisaude.com.br/quantas-arrobas-um-bezerro-engorda-por-ano/
PesoFinalEsperado

# Peso Mensal Esperado
PME = (90/12) # https://dicas.boisaude.com.br/quantas-arrobas-um-bezerro-engorda-por-ano/
PesoMensalEsperado = cattleXC_pesoMensal.groupby(by=['month'])['weightMAX'].mean().reset_index().rename(columns={'weightMAX': 'weightMEAN'})
PesoMensalEsperado['PesoMensalEsperado'] = PesoMensalEsperado['weightMEAN'] + PME
```

```python
cattleXC_pesoMensal = cattleXC_pesoMensal.merge(PesoMensalEsperado[['month', 'PesoMensalEsperado']], on='month')
```

```python
def tm_sorter(column):
    """Sort function"""
    month = ['December', 'January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October']
    correspondence = {month: order for order, month in enumerate(month)}
    return column.map(correspondence)

cattleXC_pesoMensal = cattleXC_pesoMensal.sort_values(by='month', key=tm_sorter)
```
